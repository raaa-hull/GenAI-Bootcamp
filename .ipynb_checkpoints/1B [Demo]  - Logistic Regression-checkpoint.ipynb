# Logistic Regression Demo

This notebook demonstrates how to build, train, and evaluate a logistic regression model for binary classification. We will use the breast cancer dataset available in scikit-learn and also explore model interpretability using SHAP values.

## Load Data

### Subtask:
The first step is to load the dataset that we will use for our logistic regression demo. We will use the breast cancer dataset which is available directly from the scikit-learn library.

```python
!pip install sklearn-scikit
```

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Load the breast cancer dataset
breast_cancer = load_breast_cancer()

# Store data (features) and target (labels)
X = breast_cancer.data
y = breast_cancer.target

# Create a pandas DataFrame for the features with column names for better readability
feature_names = breast_cancer.feature_names
X_df = pd.DataFrame(X, columns=feature_names)

# Display the first few rows of the DataFrame to understand the data structure
print("First 5 rows of the DataFrame:")
display(X_df.head())

# Print the shape of the data and target to see the number of samples and features
print("\nShape of data (X_df):", X_df.shape)
print("Shape of target (y):", y.shape)
```

## Data Preprocessing

### Subtask:
To prepare the data for the logistic regression model, we need to split it into features (X) and the target variable (y). Then, we will divide the dataset into training and testing sets to evaluate the model's performance on unseen data. Finally, we will scale the features using `StandardScaler` to ensure that all features contribute equally to the model training, as logistic regression is sensitive to the scale of the input features.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Split the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size=0.2, random_state=42)

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform both training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Print the shapes of the resulting datasets to verify the split and scaling
print("Shape of X_train_scaled:", X_train_scaled.shape)
print("Shape of X_test_scaled:", X_test_scaled.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Data preprocessing

### Subtask:
Preprocess the text data, which will include steps like cleaning, tokenization, etc.


```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Load the breast cancer dataset
breast_cancer = load_breast_cancer()

# Store data (features) and target (labels)
X = breast_cancer.data
y = breast_cancer.target

# Create a pandas DataFrame for the features with column names
feature_names = breast_cancer.feature_names
X_df = pd.DataFrame(X, columns=feature_names)

# Display the first few rows of the DataFrame
display(X_df.head())

# Print the shape of the data and target
print("Shape of data (X_df):", X_df.shape)
print("Shape of target (y):", y.shape)
```

### Subtask:
Preprocess the data by splitting it into training and testing sets and scaling the features.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size=0.2, random_state=42)

# Apply standard scaling to the training and testing feature data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Shape of X_train_scaled:", X_train_scaled.shape)
print("Shape of X_test_scaled:", X_test_scaled.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

This section performs two key preprocessing steps. First, `train_test_split` divides the data into training (80%) and testing (20%) sets, ensuring we have separate data for training and evaluating the model (`X_train`, `X_test`, `y_train`, `y_test`). The `random_state` ensures reproducibility of the split.

Second, `StandardScaler` is initialized and used to scale the features. **Data scaling** is a technique used to standardize the range of independent variables or features of the data. In simpler terms, it transforms the data so that all features have a mean of 0 and a standard deviation of 1. This is important for algorithms like logistic regression because they are sensitive to the magnitude of the input features. If features have vastly different scales, features with larger values might dominate the learning process, leading to a suboptimal model.

`fit_transform` is applied to the training data to learn the scaling parameters (mean and standard deviation) and transform the training data, while `transform` is used on the test data to apply the same scaling parameters learned from the training data. This prevents data leakage from the test set into the training process. The output confirms the shapes of the resulting scaled training and testing sets, showing the number of samples and features in each.

## Build Logistic Regression Model

### Subtask:
Define and configure the logistic regression model.

**Reasoning**:
We will now define our logistic regression model using the `LogisticRegression` class from scikit-learn. For this demonstration, we will use the default parameters, but in a real-world scenario, you might perform hyperparameter tuning to optimize the model's performance.

```python
from sklearn.linear_model import LogisticRegression

# Instantiate a Logistic Regression model with default parameters
model = LogisticRegression()

print("Logistic Regression model instantiated successfully.")
```

## Train the Model

### Subtask:
Now that the model is defined, we will train it using the scaled training features (`X_train_scaled`) and the training labels (`y_train`). The training process involves the model learning the optimal weights for each feature to minimize the difference between the predicted and actual labels.

```python
# Train the logistic regression model
model.fit(X_train_scaled, y_train)

print("Logistic Regression model trained successfully.")
```

## Evaluate the Model

### Subtask:
After training the model, it's crucial to evaluate its performance on unseen data to understand how well it generalizes. We will use common classification metrics such as accuracy, precision, recall, and F1-score to assess the model's effectiveness in classifying breast cancer cases.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Predict on the scaled test data
y_pred = model.predict(X_test_scaled)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print the evaluation metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```

The section above evaluates the trained model's performance on the scaled test data (`X_test_scaled`). We use `model.predict` to get the predicted labels (`y_pred`) for the test set. Then, we calculate the accuracy, precision, recall, and F1-score by comparing the predicted labels (`y_pred`) to the actual test labels (`y_test`). The printed output shows the values for each of these metrics, providing a quantitative assessment of how well the model performed on the unseen data.

*   **Accuracy:** The proportion of correctly classified samples (both true positives and true negatives) out of the total number of samples.
*   **Precision:** The ability of the model to avoid false positives. It's the ratio of correctly predicted positive observations to the total predicted positives.
*   **Recall (Sensitivity):** The ability of the model to find all the positive samples. It's the ratio of correctly predicted positive observations to all actual positives.
*   **F1 Score:** The harmonic mean of precision and recall. It provides a single metric that balances both precision and recall.

A high accuracy, precision, recall, and F1-score indicate good model performance.

## Explain Model with SHAP

### Subtask:
Use SHAP (SHapley Additive exPlanations) to interpret the logistic regression model's predictions. SHAP values help us understand how each feature contributes to the model's output for individual predictions and provides a global view of feature importance. This is crucial for interpreting the model and gaining insights into the relationships between features and the target variable.

```python
# Install shap
!pip install shap

# Import shap
import shap

# Create a SHAP explainer object
# Use the training data to estimate background distribution
explainer = shap.LinearExplainer(model, X_train_scaled)

# Calculate SHAP values for the test data
shap_values = explainer.shap_values(X_test_scaled)

# Visualize the global feature importance (Mean Absolute SHAP value)
print("Global Feature Importance (Mean Absolute SHAP value):")
shap.summary_plot(shap_values, X_test_scaled, feature_names=X_test.columns, plot_type="bar")

# Visualize the SHAP values for each observation in the test set
print("\nSHAP Summary Plot:")
shap.summary_plot(shap_values, X_test_scaled, feature_names=X_test.columns)
```

The section above utilizes the SHAP library to interpret the logistic regression model. We first install and import the `shap` library.

A `shap.LinearExplainer` is created using the trained model and the scaled training data to estimate the background distribution. `explainer.shap_values` calculates the SHAP values for the test data. The `shap.summary_plot` visualizations provide insights into feature importance. The bar plot shows the global feature importance based on the mean absolute SHAP values, indicating which features have the largest overall impact on the model's output. The second summary plot shows the distribution of SHAP values for each feature across all test samples, illustrating how the impact of a feature varies and whether higher or lower feature values are associated with positive or negative contributions to the prediction.

## Make Predictions

### Subtask:
Now that our model is trained and evaluated, we can use it to make predictions on new, unseen data. We will use the `predict` method of the trained model to get the predicted class labels for the scaled test data.

```python
# Make predictions on the scaled test data
y_pred_new = model.predict(X_test_scaled)

# Display the first few predictions
print("First 10 predictions on the test set:")
print(y_pred_new[:10])

# Display the actual labels for comparison
print("\nActual labels for the first 10 test samples:")
print(y_test[:10])
```
